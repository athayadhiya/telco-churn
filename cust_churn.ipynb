```python
import pandas as pd

data_train = pd.read_csv("E:\DS - FGA\churn\Data Train (1).csv")
data_train.head()
```

```python
data_train.info()
```

```python
data_train.isna().sum()
```

```python
data_train.duplicated().sum()
```

```python
data_train = data_train.drop(['state'], axis = 1)
```

```python
data_train = pd.get_dummies(data_train, columns = ['area_code'])
data_train.head()
```

```python
ordinal_map = {'no': 0, 'yes': 1}
ordinal_col = ['international_plan', 'voice_mail_plan', 'churn']

data_train[ordinal_col] = data_train[ordinal_col].replace(ordinal_map)
data_train.head()
```

```python
data_train.corr()
```

```python
neg_corr = ['voice_mail_plan', 'number_vmail_messages', 'total_day_calls', 'total_eve_calls', 'total_night_calls', 'total_intl_calls']

data_train = data_train.drop((neg_corr), axis = 1)
```

```python
data_train = data_train.drop(['total_day_minutes', 'total_eve_minutes', 'total_night_minutes', 'total_intl_minutes'], axis = 1)
```

```python
data_train.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (10,5))

sns.boxplot(data_train['total_day_charge'])
```

```python
plt.figure(figsize = (10,5))

sns.boxplot(data_train['total_eve_charge'])
```

```python
plt.figure(figsize = (10,5))

sns.boxplot(data_train['total_night_charge'])
```

```python
plt.figure(figsize = (10,5))

sns.boxplot(data_train['total_intl_charge'])
```

```python
print(f'Jumlah baris sebelum memfilter outlier: {len(data_train)}')

filtered_entries = np.array([True] * len(data_train))

for col in ['total_day_charge', 'total_eve_charge', 'total_night_charge', 'total_intl_charge']:
    Q1 = data_train[col].quantile(0.25)
    Q3 = data_train[col].quantile(0.75)
    
    IQR = Q3 - Q1
    
    low_limit = Q1 - (IQR * 1.5)
    high_limit = Q3 + (IQR * 1.5)

    filtered_entries = ((data_train[col] >= low_limit) & (data_train[col] <= high_limit)) & filtered_entries
    
data_train = data_train[filtered_entries]

print(f'Jumlah baris setelah memfilter outlier: {len(data_train)}')
```

```python
from sklearn.preprocessing import MinMaxScaler

data_train['total_day_charge'] = MinMaxScaler().fit_transform(data_train['total_day_charge'].values.reshape(len(data_train), 1))
data_train['total_eve_charge'] = MinMaxScaler().fit_transform(data_train['total_eve_charge'].values.reshape(len(data_train), 1))
data_train['total_night_charge'] = MinMaxScaler().fit_transform(data_train['total_night_charge'].values.reshape(len(data_train), 1))
data_train['total_intl_charge'] = MinMaxScaler().fit_transform(data_train['total_intl_charge'].values.reshape(len(data_train), 1))
```

```python
data_train.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x = data_train.drop(['churn'], axis = 1)
y = data_train['churn']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 123)
```

```python
log_reg = LogisticRegression(solver = 'lbfgs', max_iter = 500)
log_reg.fit(x_train, y_train)
```

```python
y_pred = log_reg.predict(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
y_pred2 = mnb.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score

confusion_logreg = confusion_matrix(y_test, y_pred)
accuracy_logreg = accuracy_score(y_test, y_pred)
f1_logreg = f1_score(y_test, y_pred)

print("Conf Matrix:\n", confusion_logreg)
print("Accuracy: {:.3f}".format(accuracy_logreg))
print("F1-score: {:.3f}".format(f1_logreg))
```

```python
confusion_mnb = confusion_matrix(y_test, y_pred2)
accuracy_mnb = accuracy_score(y_test, y_pred2)
f1_mnb = f1_score(y_test, y_pred2)

print("Conf Matrix:\n", confusion_mnb)
print("Accuracy: {:.3f}".format(accuracy_mnb))
print("F1-Socre: {:.3f}".format(f1_mnb))
```
